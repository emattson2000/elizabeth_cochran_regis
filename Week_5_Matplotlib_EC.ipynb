# 2. Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
df['artist'].value_counts().head(10)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

Artist = ['KG&LW', 'M. Davis', 'D. Bowie', 'T. Waits', 'B. Dylan', 'Swans', 'The Beatles', 'R.E.M.', 'Xiu Xiu', 'Aphez Twin']
Albums = [15, 14,14, 14, 13, 11, 10, 10, 10, 10]
bar_colors = ['tab:blue', 'tab:blue', 'tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue']

ax.barh(Artist, Albums, color=bar_colors)

ax.set_xlabel('Albums')
ax.set_ylabel('Artist')
ax.set_title('Artists with the Most Albums')

plt.show()
```

```python
fig.savefig('artist.jpg')
```
